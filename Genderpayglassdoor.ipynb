```python
high_salary_male = high_salary_df.Gender=='Male' 
high_salary_male 
#returns a series of boolean values
```

```python
high_salary_male_df = high_salary_df[high_salary_male] 
#passing the condition to get the dataframe for men 
high_salary_male_df
```

Out of 501, 294 are male. That is 58.68% of the data.

High salaried Female

```python
#creating a condition & passing it through the dataframe to get a new one for women  
high_salary_female = high_salary_df.Gender=='Female'   
high_salary_female_df = high_salary_df[high_salary_female] 
high_salary_female_df
```

Salary Difference & Gender

```python
high_salary_female_df.describe()
```

```python
high_salary_male_df.describe()
```

```python
high_salary_male_df.Total_Salary.mean()
```

```python
mean_salary_difference = high_salary_male_df.Total_Salary.mean() - high_salary_female_df.Total_Salary.mean()   
print('On average, men earn ${} more than women'.format(mean_salary_difference))
```

Base pay, Bonus, and Gender

```python
mean_basepay_difference = high_salary_male_df.BasePay.mean() - high_salary_female_df.BasePay.mean()  
print('Women earn ${} less than men as basepay'.format(mean_basepay_difference))
```

```python
mean_bonus_difference = high_salary_male_df.Bonus.mean() - high_salary_female_df.Bonus.mean()   

print('Men earn ${} more in bonus than women'.format(mean_bonus_difference))
```

Insights from the Data Analysis----
Total Salary Disparity:

Educational Pursuits:

Women are increasingly pursuing higher degrees, likely in hopes of achieving better remuneration in their careers.
